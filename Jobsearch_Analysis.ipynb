```python
import pandas as pd
import re
from textblob import TextBlob

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/your_file_rounded_with_LPA.csv', encoding='ISO-8859-1')

# Function to clean and correct text
def clean_text(text):
    # Ensure the value is a string before processing
    if isinstance(text, str):
        # Remove special characters and extra spaces
        text = re.sub(r'[^A-Za-z0-9\s]', '', text)  # Remove non-alphanumeric characters except spaces

        # Correct spelling using TextBlob
        corrected_text = str(TextBlob(text).correct())

        # Capitalize the first letter of each word
        corrected_text = corrected_text.title()
        return corrected_text
    return text  # If it's not a string, return it as is (e.g., NaN or numeric)

# Apply the function to the 'Position' column
if 'Position' in df.columns:
    df['Position'] = df['Position'].apply(clean_text)

# Apply the function to the 'Education' column
if 'Education' in df.columns:
    df['Education'] = df['Education'].apply(clean_text)

# Apply the function to the 'Location' column
if 'Location' in df.columns:
    df['Location'] = df['Location'].apply(lambda loc: str(loc).strip().title())

# Save the cleaned DataFrame to a new CSV
df.to_csv('position_salary_cleaned_education_location.csv', index=False)

print("Position, Education, and Location columns cleaned and saved successfully!")
```

```python
import spacy
import pandas as pd

# Load the spaCy model for English
nlp = spacy.load("en_core_web_sm")

# Define a function to tokenize text and include numeric fields explicitly
def tokenize_text_with_numbers(row, fields_to_tokenize):
    tokens = []

    # Process each field
    for field in fields_to_tokenize:
        if field in row and pd.notnull(row[field]):
            value = str(row[field])
            if field in ['Salary','Experience (Years)']:
                # Add numeric value with descriptive token
                tokens.append(f"{value}")
            else:
                # Tokenize text fields using spaCy
                doc = nlp(value.lower())
                tokens.extend([token.text for token in doc if token.is_alpha])

    return tokens

# Load the CSV file
df = pd.read_csv('/content/your_file_rounded_with_LPA.csv', encoding='ISO-8859-1')

# Specify the fields to tokenize
fields_to_tokenize = ['Position', 'Education', 'Location', 'Salary', 'Salary_in_LPA', 'Experience (Years)']

# Apply tokenization to each row and create a new column
df['Combined_Tokens'] = df.apply(lambda row: tokenize_text_with_numbers(row, fields_to_tokenize), axis=1)

# Save the updated DataFrame to a new CSV
df.to_csv('position_salary_combined_tokenized_with_numbers2.csv', index=False)

print("Tokenization completed. Tokens added to the 'Combined_Tokens' column.")
```

```python
import os
import csv
import spacy
from whoosh.index import create_in, open_dir
from whoosh.fields import Schema, TEXT, ID, NUMERIC
from whoosh.qparser import MultifieldParser
from fuzzywuzzy import fuzz

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define schema for the index
schema = Schema(
    id=ID(stored=True, unique=True),
    position=TEXT(stored=True),
    combined_tokens=TEXT(stored=True),
    location=TEXT(stored=True),
    education=TEXT(stored=True),
    experience=NUMERIC(stored=True),
    salary=NUMERIC(stored=True),
    skills=TEXT(stored=True)
)

# Define index directory
index_dir = "job_index"

# Create or open the index
if os.path.exists(index_dir) and len(os.listdir(index_dir)) > 0:
    ix = open_dir(index_dir)
else:
    os.mkdir(index_dir)
    ix = create_in(index_dir, schema)

# Indexing data
csv_file = "/content/position_salary_combined_tokenized_with_numbers.csv"
with ix.writer() as writer:
    with open(csv_file, "r", encoding="ISO-8859-1") as f:
        reader = csv.DictReader(f)
        for i, row in enumerate(reader):
            writer.add_document(
                id=str(i),
                position=row.get("Position", ""),
                combined_tokens=row.get("Combined_Tokens", ""),
                location=row.get("Location", ""),
                education=row.get("Education", ""),
                experience=int(row.get("Experience (Years)", 0)),
                salary=int(row.get("Salary", 0)),
                skills=row.get("Skills", "")
            )

print("Indexing completed.")

# Function to preprocess query into tokens
def preprocess_query(query):
    doc = nlp(query)
    tokens = [token.text.lower() for token in doc if not token.is_punct and not token.is_stop]
    return tokens

# Perform fuzzy search
def fuzzy_search(query_string, top_n=10):
    user_tokens = preprocess_query(query_string)  # Tokenize user query
    query_combined = " ".join(user_tokens)        # Combine tokens into a single string for matching

    with ix.searcher() as searcher:
        results = []

        # Iterate through indexed documents
        for doc in searcher.documents():
            combined_tokens = doc["combined_tokens"].lower()
            fuzzy_score = fuzz.token_sort_ratio(query_combined, combined_tokens)

            results.append({
                "position": doc["position"],
                "location": doc["location"],
                "education": doc["education"],
                "experience": doc["experience"],
                "salary": doc["salary"],
                "skills": doc["skills"],
                "combined_tokens": combined_tokens,
                "score": fuzzy_score
            })

        # Sort results by fuzzy score in descending order
        sorted_results = sorted(results, key=lambda x: x["score"], reverse=True)[:top_n]

        # Display results
        if not sorted_results:
            print("No matching jobs found.")
        else:
            for result in sorted_results:
                print("Position:", result["position"])
                print("Location:", result["location"])
                print("Education:", result["education"])
                print("Experience (Years):", result["experience"])
                print("Salary:", result["salary"])
                print("Skills:", result["skills"])
                print("Combined Tokens:", result["combined_tokens"])
                print("Fuzzy Score:", result["score"])
                print("-" * 40)

# Example query
query = "jobs in Ghaziabad with 2000000"
fuzzy_search(query)
```

```python
import os
import csv
import spacy
from whoosh.index import create_in, open_dir
from whoosh.fields import Schema, TEXT, ID, NUMERIC
from fuzzywuzzy import fuzz

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define schema for the index
schema = Schema(
    id=ID(stored=True, unique=True),
    position=TEXT(stored=True),
    combined_tokens=TEXT(stored=True),
    location=TEXT(stored=True),
    education=TEXT(stored=True),
    experience=NUMERIC(stored=True),
    salary=NUMERIC(stored=True),
    skills=TEXT(stored=True)
)

# Define index directory
index_dir = "job_index"

# Create or open the index
if os.path.exists(index_dir) and len(os.listdir(index_dir)) > 0:
    ix = open_dir(index_dir)
else:
    os.mkdir(index_dir)
    ix = create_in(index_dir, schema)

# Indexing data
csv_file = "/content/position_salary_combined_tokenized_with_numbers2.csv"
with ix.writer() as writer:
    with open(csv_file, "r", encoding="ISO-8859-1") as f:
        reader = csv.DictReader(f)
        for i, row in enumerate(reader):
            writer.add_document(
                id=str(i),
                position=row.get("Position", ""),
                combined_tokens=row.get("Combined_Tokens", ""),
                location=row.get("Location", ""),
                education=row.get("Education", ""),
                experience=int(row.get("Experience (Years)", 0)),
                salary=int(row.get("Salary", 0)),
                skills=row.get("Skills", "")
            )

print("Indexing completed.")

# Function to extract keywords from text
def extract_keywords(text):
    doc = nlp(text)
    keywords = [token.text.lower() for token in doc if token.pos_ in {"NOUN", "PROPN", "ADJ"}]
    return set(keywords)

# Perform keyword-based search
def keyword_relevancy_search(query_string, top_n=10):
    user_keywords = extract_keywords(query_string)  # Extract keywords from user query

    with ix.searcher() as searcher:
        results = []

        # Iterate through indexed documents
        for doc in searcher.documents():
            combined_tokens = doc["combined_tokens"].lower()
            combined_keywords = extract_keywords(combined_tokens)

            # Calculate relevancy as the number of matching keywords
            matching_keywords = user_keywords.intersection(combined_keywords)
            matching_count = len(matching_keywords)

            # Calculate fuzzy score for added context
            fuzzy_score = fuzz.token_sort_ratio(query_string.lower(), combined_tokens)

            results.append({
                "position": doc["position"],
                "location": doc["location"],
                "education": doc["education"],
                "experience": doc["experience"],
                "salary": doc["salary"],
                "skills": doc["skills"],
                "combined_tokens": combined_tokens,
                "matching_count": matching_count,
                "fuzzy_score": fuzzy_score,
                "matching_keywords": list(matching_keywords)
            })

        # Sort results by matching keyword count, then fuzzy score
        sorted_results = sorted(
            results,
            key=lambda x: (x["matching_count"], x["fuzzy_score"]),
            reverse=True
        )[:top_n]

        # Display results
        if not sorted_results:
            print("No matching jobs found.")
        else:
            print(f"Top {top_n} Matching Results:\n")
            for result in sorted_results:
                print("Position:", result["position"])
                print("Location:", result["location"])
                print("Education:", result["education"])
                print("Experience (Years):", result["experience"])
                print("Salary:", result["salary"])
                print("Skills:", result["skills"])
                print("Combined Tokens:", result["combined_tokens"])
                print("Matching Keywords:", result["matching_keywords"])
                print("Matching Keywords Count:", result["matching_count"])
                print("Fuzzy Score:", result["fuzzy_score"])
                print("-" * 40)

# Example query
query = "jobs for engineer QA in Bangalore with 11 years of experience and 21LPA"
keyword_relevancy_search(query)
```
